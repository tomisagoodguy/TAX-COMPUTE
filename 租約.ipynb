```python

import openpyxl
from openpyxl import Workbook
from openpyxl.styles import Font, Alignment, PatternFill, Border, Side, Protection
from openpyxl.worksheet.datavalidation import DataValidation
from openpyxl.worksheet.protection import SheetProtection
from openpyxl.utils import get_column_letter
import datetime

def generate_final_excel():
    # ===================== 0. åˆå§‹åŒ– =====================
    wb = Workbook()
    wb.remove(wb.active)
    ws = wb.create_sheet("äº¤å±‹åˆ†ç®—(å®šç¨¿ç‰ˆ)", 0)
    ws.sheet_view.showGridLines = False

    # ===================== 1. ç¾å­¸è¨­å®š (High Contrast) =====================
    # æ¨™é¡Œå€
    COLOR_HEADER_BG = "1E293B"    # Slate 800 (æ·±è‰²å°ˆæ¥­)
    COLOR_HEADER_FONT = "FFFFFF"  # ç™½å­—
    
    # è¼¸å…¥å€ (Input Zone) - ä¿æŒæ¥µç°¡ä¹¾æ·¨ï¼Œè®“ä½¿ç”¨è€…å°ˆæ³¨å¡«å¯«
    COLOR_INPUT_BG = "FFFFFF"     # ç´”ç™½
    COLOR_INPUT_HEADER = "64748B" # Slate 500 (æ¨™é¡Œåº•è‰² - ä¸­æ€§ç°)
    COLOR_INPUT_TXT = "FFFFFF"
    
    # è³£æ–¹å€ (Seller Zone) - æš–è‰²ç³»è­¦ç¤º
    COLOR_SELLER_BG = "FFF1F2"    # Rose 50 (åº•è‰²)
    COLOR_SELLER_HEADER = "E11D48" # Rose 600 (æ¨™é¡Œåº•è‰²)
    COLOR_SELLER_TXT = "881337"   # Rose 900 (æ–‡å­—)
    
    # è²·æ–¹å€ (Buyer Zone) - å†·è‰²ç³»å€éš”
    COLOR_BUYER_BG = "ECFDF5"     # Emerald 50 (åº•è‰²)
    COLOR_BUYER_HEADER = "059669" # Emerald 600 (æ¨™é¡Œåº•è‰²)
    COLOR_BUYER_TXT = "064E3B"    # Emerald 900 (æ–‡å­—)
    
    # çµç®—å€
    COLOR_FINAL_BG = "FFEDD5"     # Orange 100
    
    # é‚Šæ¡†
    COLOR_BORDER = "CBD5E1"       # Slate 300
    COLOR_BORDER_THICK = "334155" # Slate 700 (å€å¡Šåˆ†éš”ç·š)

    font_title = Font(name="å¾®è»Ÿæ­£é»‘é«”", size=24, bold=True, color="1E293B")
    font_header_base = Font(name="å¾®è»Ÿæ­£é»‘é«”", size=10, bold=True, color="FFFFFF")
    
    font_normal = Font(name="å¾®è»Ÿæ­£é»‘é«”", size=10, color="334155")
    font_input = Font(name="å¾®è»Ÿæ­£é»‘é«”", size=11, color="000000") # è¼¸å…¥æ–‡å­—æ·±é»‘
    
    font_seller = Font(name="Arial", size=10, bold=True, color=COLOR_SELLER_TXT)
    font_buyer = Font(name="Arial", size=10, bold=True, color=COLOR_BUYER_TXT)
    
    font_big_res = Font(name="Arial", size=20, bold=True, color="1E293B")
    font_hint = Font(name="å¾®è»Ÿæ­£é»‘é«”", size=9, color="64748B")

    # æ¨™æº–é‚Šæ¡†
    border_std = Border(
        left=Side(style='thin', color=COLOR_BORDER),
        right=Side(style='thin', color=COLOR_BORDER),
        top=Side(style='thin', color=COLOR_BORDER),
        bottom=Side(style='thin', color=COLOR_BORDER)
    )
    
    # ç²—é‚Šæ¡† (ç”¨æ–¼å€éš”è¼¸å…¥å€èˆ‡è¨ˆç®—å€)
    border_thick_right = Border(
        left=Side(style='thin', color=COLOR_BORDER),
        right=Side(style='medium', color=COLOR_BORDER_THICK), # å³å´åŠ ç²—
        top=Side(style='thin', color=COLOR_BORDER),
        bottom=Side(style='thin', color=COLOR_BORDER)
    )
    
    border_thick_bottom = Border(bottom=Side(style='medium', color="1E293B"))

    align_c = Alignment(horizontal="center", vertical="center", wrap_text=True)
    align_r = Alignment(horizontal="right", vertical="center")
    align_l = Alignment(horizontal="left", vertical="center", wrap_text=True)

    def set_cell(cell_ref, val, font=font_normal, fill=None, align=align_c, border=border_std, fmt=None):
        c = ws[cell_ref]
        c.value = val
        c.font = font
        c.alignment = align
        if fill: c.fill = PatternFill("solid", fgColor=fill)
        if border: c.border = border
        if fmt: c.number_format = fmt
        return c

    # ===================== 2. æ¬„ä½è¦åŠƒ =====================
    # A:æˆ¿è™Ÿ, B:ä»˜æ¬¾æ—¥, C:èµ·, D:è¿„, E:æœˆç§Ÿ, F:æŠ¼é‡‘, G:ç¹³è‡³ (è¼¸å…¥å€ Gæ¬„å³å´è¦åŠ ç²—ç·š)
    # H:è³£è£œè²·(è©³), I:è³£è£œè²·(é‡‘)
    # J:è²·è£œè³£(è©³), K:è²·è£œè³£(é‡‘)
    cols = [
        ("A", 10), ("B", 10), 
        ("C", 10), ("D", 10), 
        ("E", 12), ("F", 12), ("G", 12), 
        ("H", 24), ("I", 12), # èªªæ˜æ¬„åŠ å¯¬
        ("J", 24), ("K", 12)  
    ]
    for col_char, width in cols:
        ws.column_dimensions[col_char].width = width

    # ===================== 3. å„€è¡¨æ¿é ­éƒ¨ =====================
    ws.merge_cells("A1:K1")
    set_cell("A1", "ä¸å‹•ç”¢äº¤å±‹ Â· ç§Ÿé‡‘åˆ†ç®—çµç®—è¡¨ (å®šç¨¿ç‰ˆ)", font=font_title, align=align_l, border=None)
    
    for col in range(1, 12):
        ws.cell(2, col).border = border_thick_bottom

    # --- é—œéµåƒæ•¸ ---
    set_cell("A3", "äº¤å±‹æ—¥æœŸ (7ç¢¼)", font=font_header_base, fill=COLOR_HEADER_BG)
    set_cell("B3", 1140315, font=font_input, fill=COLOR_INPUT_BG, fmt="0") 
    ws.merge_cells("C3:E3")
    verify_formula = f"""=IF(B3="","è«‹è¼¸å…¥æ—¥æœŸ", "ğŸ“… " & TEXT(DATE(1911+LEFT(B3,3),MID(B3,4,2),RIGHT(B3,2)),"yyyyå¹´mmæœˆddæ—¥"))"""
    set_cell("C3", verify_formula, font=font_hint, align=align_l, border=None)

    set_cell("A4", "äº¤å±‹æ—¥ç®—èª°çš„", font=font_header_base, fill=COLOR_HEADER_BG)
    set_cell("B4", "æ­¸è²·æ–¹", font=font_input, fill=COLOR_INPUT_BG)
    ws.merge_cells("C4:E4")
    set_cell("C4", '="âš–ï¸ æ¬Šç›Šåˆ‡å‰²ï¼šäº¤å±‹æ—¥ç•¶å¤©æ­¸ " & RIGHT(B4,2)', font=font_hint, align=align_l, border=None)

    dv_owner = DataValidation(type="list", formula1='"æ­¸è²·æ–¹,æ­¸è³£æ–¹"', allow_blank=False)
    ws.add_data_validation(dv_owner)
    dv_owner.add("B4:B4")

    set_cell("A5", "è¨ˆç®—åŸºæº–", font=font_header_base, fill=COLOR_HEADER_BG)
    set_cell("B5", "ç•¶æœˆå¯¦éš›å¤©æ•¸", font=font_normal, fill="F1F5F9")
    ws.merge_cells("C5:F5")
    days_in_month_formula = f"""=IF(B3="","", "ğŸ“Š åˆ†æ¯ï¼šä¾äº¤å±‹ç•¶æœˆç¸½å¤©æ•¸ (" & DAY(EOMONTH(DATE(1911+LEFT(B3,3),MID(B3,4,2),RIGHT(B3,2)),0)) & "å¤©) è¨ˆç®—æ—¥ç§Ÿé‡‘")"""
    set_cell("C5", days_in_month_formula, font=font_hint, align=align_l, border=None)

    # ===================== 4. æ ¸å¿ƒæ•¸æ“šè¡¨ =====================
    headers = [
        "æˆ¿è™Ÿ/ç§Ÿå®¢", "ä»˜æ¬¾æ—¥\n(åƒè€ƒç”¨)", 
        "ç§ŸæœŸèµ·", "ç§ŸæœŸè¨–", 
        "æœˆç§Ÿé‡‘", "æŠ¼é‡‘", "å·²ç¹³è‡³", 
        "è³£æ–¹è£œè²¼è²·æ–¹\n(è¨ˆç®—è©³æƒ…)", "è£œè²¼é‡‘é¡", 
        "è²·æ–¹è£œè²¼è³£æ–¹\n(è¨ˆç®—è©³æƒ…)", "è£œè²¼é‡‘é¡"
    ]
    
    for i, h in enumerate(headers):
        col = get_column_letter(i+1)
        
        # é è¨­æ¨£å¼ (Input Zone)
        bg_color = "64748B" # Slate 500
        txt_color = "FFFFFF"
        
        # ç‰¹æ®Šå€åŸŸæ¨£å¼
        if "åƒè€ƒ" in h: 
            bg_color = "94A3B8" # Slate 400 (Lighter)
        
        if "è³£æ–¹" in h: 
            bg_color = COLOR_SELLER_HEADER
        
        if "è²·æ–¹" in h: 
            bg_color = COLOR_BUYER_HEADER
            
        # é‚Šæ¡†è™•ç†ï¼šGæ¬„å³å´è¦ç²—
        cell_border = border_std
        if col == "G":
            cell_border = border_thick_right
            
        set_cell(f"{col}7", h, font=font_header_base, fill=bg_color, border=cell_border)

    start_row = 8
    end_row = 27

    # ä»˜æ¬¾æ—¥é¸å–®
    days_list = [str(i) for i in range(1, 32)] + ["æœˆåº•"]
    formula_str = f'"{",".join(days_list)}"'
    dv_payment = DataValidation(type="list", formula1=formula_str, allow_blank=True)
    ws.add_data_validation(dv_payment)
    dv_payment.add(f"B{start_row}:B{end_row}")

    for r in range(start_row, end_row + 1):
        row_str = str(r)
        
        # è¼¸å…¥å€ (A-G)
        # ç‰¹åˆ¥è™•ç† G æ¬„é‚Šæ¡†
        for c_char in ["A","B","C","D","E","F"]:
            set_cell(f"{c_char}{row_str}", "", fill=COLOR_INPUT_BG)
        set_cell(f"G{row_str}", "", fill=COLOR_INPUT_BG, border=border_thick_right) # Gæ¬„åŠ ç²—å³æ¡†

        # æ ¼å¼è¨­å®š
        ws[f"B{row_str}"].number_format = "@"
        ws[f"C{row_str}"].number_format = "0"
        ws[f"D{row_str}"].number_format = "0"
        ws[f"E{row_str}"].number_format = "#,##0"
        ws[f"F{row_str}"].number_format = "#,##0"
        ws[f"G{row_str}"].number_format = "0"

        # --- è®Šæ•¸å®šç¾© ---
        safe_b3 = f"IF($B$3=\"\", 1140101, $B$3)"
        d_handover = f"DATE(1911+LEFT({safe_b3},3),MID({safe_b3},4,2),RIGHT({safe_b3},2))"
        d_m_start = f"DATE(YEAR({d_handover}),MONTH({d_handover}),1)"
        d_m_end = f"EOMONTH({d_handover},0)"
        
        safe_g = f"IF(G{row_str}=\"\", {safe_b3}, G{row_str})"
        d_paid = f"DATE(1911+LEFT({safe_g},3),MID({safe_g},4,2),RIGHT({safe_g},2))"
        
        d_cut = f"IF($B$4=\"æ­¸è²·æ–¹\", {d_handover}, {d_handover}+1)"
        denom_days = f"DAY({d_m_end})"

        # --- æ¬Šç›Šåˆ¤æ–· ---
        entitled_days = f"MAX(0, {d_cut} - {d_m_start})"
        collected_days = f"MAX(0, MIN({d_paid}, {d_m_end}) - {d_m_start} + 1)"
        net_days = f"({collected_days} - {entitled_days})"
        
        is_valid = f'AND(A{row_str}<>"", E{row_str}<>"", G{row_str}<>"")'

        # --- H: è³£æ–¹è£œè²¼è²·æ–¹ (èªªæ˜) ---
        s_date_start = d_cut
        s_date_end = f"MIN({d_paid}, {d_m_end})"
        
        # å„ªåŒ–é¡¯ç¤ºæ ¼å¼ï¼šä½¿ç”¨æ‹¬è™ŸåŒ…è¦†åˆ†æ¯ (Days/Denom)
        # CHAR(10) æ˜¯æ›è¡Œç¬¦
        s_desc_formula = f"""=IF({is_valid}, IF({net_days}>0, TEXT({s_date_start}, "mm/dd") & "~" & TEXT({s_date_end}, "mm/dd") & " (" & {net_days} & "å¤©)" & CHAR(10) & "$" & TEXT(E{row_str}, "#,##0") & "Ã—(" & {net_days} & "/" & {denom_days} & ")", ""), "")"""
        set_cell(f"H{row_str}", s_desc_formula, fill=COLOR_SELLER_BG, font=font_seller)
        ws[f"H{row_str}"].alignment = Alignment(wrap_text=True, horizontal="center", vertical="center")

        # --- I: è³£æ–¹è£œè²¼è²·æ–¹ (é‡‘é¡) ---
        s_amt_formula = f"""=IF({is_valid}, IF({net_days}>0, ROUND(E{row_str}/{denom_days}*{net_days}, 0), 0), "")"""
        set_cell(f"I{row_str}", s_amt_formula, fill=COLOR_SELLER_BG, fmt="#,##0", font=font_seller)

        # --- J: è²·æ–¹è£œè²¼è³£æ–¹ (èªªæ˜) ---
        b_date_start = f"MAX({d_m_start}, {d_paid}+1)"
        b_date_end = f"{d_cut}-1"
        
        # å„ªåŒ–é¡¯ç¤ºæ ¼å¼
        b_desc_formula = f"""=IF({is_valid}, IF({net_days}<0, TEXT({b_date_start}, "mm/dd") & "~" & TEXT({b_date_end}, "mm/dd") & " (" & ABS({net_days}) & "å¤©)" & CHAR(10) & "$" & TEXT(E{row_str}, "#,##0") & "Ã—(" & ABS({net_days}) & "/" & {denom_days} & ")", ""), "")"""
        set_cell(f"J{row_str}", b_desc_formula, fill=COLOR_BUYER_BG, font=font_buyer)
        ws[f"J{row_str}"].alignment = Alignment(wrap_text=True, horizontal="center", vertical="center")

        # --- K: è²·æ–¹è£œè²¼è³£æ–¹ (é‡‘é¡) ---
        b_amt_formula = f"""=IF({is_valid}, IF({net_days}<0, ROUND(E{row_str}/{denom_days}*ABS({net_days}), 0), 0), "")"""
        set_cell(f"K{row_str}", b_amt_formula, fill=COLOR_BUYER_BG, fmt="#,##0", font=font_buyer)

    # ç¯„ä¾‹
    set_cell("A8", "301 (æ¸¬è©¦)", fill=COLOR_INPUT_BG)
    set_cell("B8", "æœˆåº•", fill=COLOR_INPUT_BG, align=align_c)
    set_cell("C8", 1130101, fill=COLOR_INPUT_BG, fmt="0")
    set_cell("D8", 1141231, fill=COLOR_INPUT_BG, fmt="0")
    set_cell("E8", 15000, fill=COLOR_INPUT_BG, fmt="#,##0")
    set_cell("F8", 30000, fill=COLOR_INPUT_BG, fmt="#,##0")
    set_cell("G8", 1140228, fill=COLOR_INPUT_BG, fmt="0", border=border_thick_right) 

    # ===================== 5. çµç®—å€ =====================
    sum_row = end_row + 2
    
    # è³£æ–¹æ‡‰ä»˜ (Debits) - è¦–è¦ºå¼·åŒ–
    ws.merge_cells(f"A{sum_row}:E{sum_row}")
    set_cell(f"A{sum_row}", "è³£æ–¹æ‡‰ä»˜é …ç›® (Debits)", font=font_header_base, fill=COLOR_SELLER_HEADER, align=align_l)
    
    set_cell(f"A{sum_row+1}", "1. æŠ¼é‡‘ç§»è½‰", align=align_l, border=None)
    set_cell(f"E{sum_row+1}", f"=SUM(F{start_row}:F{end_row})", fmt="#,##0", font=font_seller, align=align_r)
    
    set_cell(f"A{sum_row+2}", "2. è³£æ–¹è£œè²¼è²·æ–¹ (æº¢æ”¶ç§Ÿé‡‘)", align=align_l, border=None)
    set_cell(f"E{sum_row+2}", f"=SUM(I{start_row}:I{end_row})", fmt="#,##0", font=font_seller, align=align_r)

    # è²·æ–¹æ‡‰ä»˜ (Credits) - è¦–è¦ºå¼·åŒ–
    ws.merge_cells(f"G{sum_row}:K{sum_row}")
    set_cell(f"G{sum_row}", "è²·æ–¹æ‡‰ä»˜é …ç›® (Credits)", font=font_header_base, fill=COLOR_BUYER_HEADER, align=align_l)
    
    set_cell(f"G{sum_row+1}", "1. è²·æ–¹è£œè²¼è³£æ–¹ (çŸ­æ”¶ç§Ÿé‡‘)", align=align_l, border=None)
    set_cell(f"K{sum_row+1}", f"=SUM(K{start_row}:K{end_row})", fmt="#,##0", font=font_buyer, align=align_r)
    
    set_cell(f"G{sum_row+2}", "2. å…¶ä»–ä»£å¢Šè²»ç”¨ (ç®¡ç†è²»/ç¨…)", align=align_l, border=None)
    set_cell(f"K{sum_row+2}", 0, fmt="#,##0", font=font_buyer, fill=COLOR_INPUT_BG, align=align_r) 

    # Net
    final_row = sum_row + 4
    ws.merge_cells(f"A{final_row}:D{final_row}")
    ws.merge_cells(f"E{final_row}:K{final_row}")

    net_val = f"E{final_row}"
    ws[net_val] = f"=(E{sum_row+1}+E{sum_row+2}) - (K{sum_row+1}+K{sum_row+2})"
    ws[net_val].font = font_big_res
    ws[net_val].alignment = align_c
    ws[net_val].number_format = "#,##0"
    ws[net_val].fill = PatternFill("solid", fgColor=COLOR_FINAL_BG)
    
    text_formula = f"""=IF({net_val}>=0, "æœ€çµ‚çµç®—ï¼šè³£æ–¹ æ‡‰æ‰¾è£œ è²·æ–¹", "æœ€çµ‚çµç®—ï¼šè²·æ–¹ æ‡‰æ‰¾è£œ è³£æ–¹")"""
    set_cell(f"A{final_row}", text_formula, font=Font(size=14, bold=True, color="1E293B"), align=align_r, border=None)

    # ===================== 6. ä¿è­· =====================
    ws.protection = SheetProtection(sheet=True, formatCells=False, formatColumns=False)
    
    unlock_ranges = [
        "B3:B4", 
        f"A{start_row}:G{end_row}", 
        f"K{sum_row+2}:K{sum_row+2}" 
    ]
    
    for rng in unlock_ranges:
        for row in ws[rng]:
            for cell in row:
                cell.protection = Protection(locked=False)

    filename = f"äº¤å±‹åˆ†ç®—_å®šç¨¿ç‰ˆ_{datetime.datetime.now().strftime('%H%M%S')}.xlsx"
    wb.save(filename)
    return filename

if __name__ == "__main__":
    print("ç”Ÿæˆå®šç¨¿ç‰ˆ Excel...")
    f = generate_final_excel()
    print(f"å®Œæˆï¼æª”æ¡ˆï¼š{f}")
```
